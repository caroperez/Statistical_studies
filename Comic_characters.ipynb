### Number of new appearances over time

```python
# magic command for ploting
%matplotlib inline


# data for plot
dates = chart_descriptions['Year'].to_frame()
df = dates['Year'].value_counts().to_frame()

# plot definition
plt.figure(figsize = (8, 6), dpi = 90)
plt.plot(df.sort_index(), color="Slateblue", alpha=0.6)

# plot names axes
plt.xlabel('Year')
plt.ylabel('Number of first appearances')
plt.title('Number of appearance along of time')

# show
plt.show()
```
